```python
import pandas as pd
import matplotlib as plt

%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
data = pd.read_csv('../work/nba_18_19_pg-Copy1.csv', delimiter=',')
```

```python
data
```

```python
subset = data[data['FT%']>.9]
subset = subset.sort_values(by='Age', ascending=True)
```

```python
plot = subset['Age'].plot(kind='bar',figsize=(20,5), label='nba players age', xlabel="Players", ylabel="age", legend=True, color='g', title='90%+ Free Throw Shooters by Age, 2018-2019 Season')
plot.set_xticklabels(subset['Player']);
```

```python
# My plot shows the age of NBA players that had a free throw rate above 90% in the years 2018-2019
```

```python
# I decided to make a bar graph because I figured that it would be the clearest way to visually compare the players' ascending age. I specifically avoided a pie chart because I know y'all dislike them and when I tried a line graph, the information looked chaotic.
```
